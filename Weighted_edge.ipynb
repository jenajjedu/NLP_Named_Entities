```python
def save_graph(G, filename):
    print()
    nx.write_graphml(G, (f'{os.getcwd()}/{filename}.graphml'))
    print(f'save a serialization of the {filename} graph in graphml format to be read by the Cytoscape app')
    nx.draw_networkx(G)
    print(f'save a plot of the {filename} graph in matplotlib and export as a png')
    plt.plot(marker=".", markersize=1)
    plt.savefig(f'{os.getcwd()}/{filename}.png')
    # The file collab.graphml can be viewed using Cytoscape.
```
